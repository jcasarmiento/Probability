```python
import json
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
def change_accents(words): #to normalize the number of letters used  
    words=words.replace("á","a")
    words=words.replace("é","e")
    words=words.replace("í","i")
    words=words.replace("ó","o")
    words=words.replace("ú","u")
    words=words.replace("ü","u")
    words=words.replace("ç","c")
    words=words.replace("ñ","n")
    return words
```

```python
def readfile(file):
    f = open(file, "r",encoding='utf8')
    text=""
    for line in f:
        text=text+" "+line
    f.close()
    return text
```

```python
#spanish="El desconocido llegó un día huracanado de primeros de febrero, abriéndose paso a través de un viento cortante y de una densa nevada, la última del año. El desconocido llegó a pie desde la estación del ferrocarril de Bramblehurst. Llevaba en la mano bien enguantada una pequeña maleta negra. Iba envuelto de los pies a la cabeza, el ala de su sombrero de fieltro le tapaba todo el rostro y sólo dejaba al descubierto la punta de su nariz. La nieve se había ido acumulando sobre sus hombros y sobre la pechera de su atuendo y había formado una capa blanca en la parte superior de su carga. Más muerto que vivo, entró tambaleándose en la fonda Coach and Horses y, después de soltar su maleta, gritó: «¡Un fuego, por caridad! ¡Una habitación con un fuego!» Dio unos golpes en el suelo y se sacudió la nieve junto a la barra. Después siguió a la señora Hall hasta el salón para concertar el precio. Sin más presentaciones, una rápida conformidad y un par de soberanos sobre la mesa, se alojó en la posada."
#H.G. Wells. El Hombre Invisible. From: https://infolibros.org/pdfview/6094-el-hombre-invisible-herbert-george-wells/
spanish=readfile("spanish.txt")
```

```python
#english="No one would have believed in the last years of the nineteenth century that this world was being watched keenly and closely by intelligences greater than man's and yet as mortal as his own; that as men busied themselves about their various concerns they were scrutinised and studied, perhaps almost as narrowly as a man with a microscope might scrutinise the transient creatures that swarm and multiply in a drop of water. With infinite complacency men went to and fro over this globe about their little affairs, serene in their assurance of their empire over matter. It is possible that the infusoria under the microscope do the same. No one gave a thought to the older worlds of space as sources of human danger, or thought of them only to dismiss the idea of life upon them as impossible or improbable. It is curious to recall some of the mental habits of those departed days. At most terrestrial men fancied there might be other men upon Mars, perhaps inferior to themselves and ready to welcome a missionary enterprise. Yet across the gulf of space, minds that are to our minds as ours are to those of the beasts that perish, intellects vast and cool and unsympathetic, regarded this earth with envious eyes, and slowly and surely drew their plans against us. And early in the twentieth century came the great disillusionment."
#H.G. Wells. The War of the Worlds. InfoBooks.org. From: https://www.infobooks.org/books/the-war-of-the-worlds-h-g-wells.pdf
english=readfile("english.txt")
```

```python
alphabet="abcdefghijklmnopqrstuvwxyz"

def count_chars(words):
    counter={}
    for letter in alphabet:
        counter[letter]=words.count(letter)
    return counter

counts_es=count_chars(change_accents(spanish.lower()))
total_es=sum(counts_es.values())
counts_en=count_chars(change_accents(english.lower()))
total_en=sum(counts_en.values())
```

```python
plt.figure(figsize=(5,1.5))
plt.bar(range(len(counts_es)), counts_es.values(), align='center', color="orange")
plt.xticks(range(len(counts_es)), list(counts_es.keys()))
plt.ylabel("Counts")
plt.xlabel("Character")
plt.title("Histogram of characters in El Hombre Invisible")
plt.show()
```

```python
plt.figure(figsize=(5,1.5))
counts_es_normal={key: value/total_es for key, value in counts_es.items()}
total=sum(counts_es_normal.values())
plt.bar(range(len(counts_es)), counts_es_normal.values(), align='center', color="orange")
plt.xticks(range(len(counts_es)), list(counts_es.keys()))
plt.ylabel("Probability $P(X=x)$")
plt.xlabel("Character $x$")
plt.title("Empirical PMF of El Hombre Invisible")
plt.show()
```

```python
def likelihood(new_seq,p):
    res=1.0
    for char in new_seq:
        if char in alphabet:
            res=res*p[char]
            #print(res)
    return res

def log_likelihood(new_seq,p):
    res=0.0
    for char in new_seq:
        if char in alphabet:
            res=res+np.log(p[char])
            #print(char,"log_likelihood",res,"P(X)",p[char],sep=",")
    return res      
```

```python
def predict(words):
    es=log_likelihood(change_accents(words.lower()),counts_es_normal)
    #print("es",es)
    en=log_likelihood(change_accents(words.lower()),counts_en_normal)
    #print("en",en)
    if es>en:
        return "es"
    else:
        return "en"
```
